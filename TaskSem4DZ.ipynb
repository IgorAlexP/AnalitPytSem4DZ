2.1. Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение CSV-файла
data = pd.read_csv('kc-house-data.csv')

# Подсчет количества домов с и без вида на набережную
waterfront_counts = data['waterfront'].value_counts()

# Построение графика распределения
labels = ['Нет вида на набережную', 'Есть вид на набережную']
plt.bar(labels, waterfront_counts)
plt.xlabel('Наличие вида на набережную')
plt.ylabel('Количество домов')
plt.title('Распределение домов от наличия вида на набережную')
plt.show()
```

по этому граффику мы можем сделать вывод сколько домов имеет вид на набережную,а сколько нет. Делаем вывод, что количество домов без вида на набережную значительно преобладает над количеством домов с видом на набережную.

### 2.2. Изучите распределение этажей домов

```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение CSV-файла
data = pd.read_csv('kc-house-data.csv')

# Подсчет количества домов по количеству этажей
floors_counts = data['floors'].value_counts()

# Построение графика распределения
plt.bar(floors_counts.index, floors_counts.values)
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.title('Распределение этажей домов')
plt.show()
```

Анализируя этот график, мы можем сделать выводы о распределении этажей домов. Например, можно увидеть, сколько домов имеют определенное количество этажей и определить наиболее распространенное количество этажей. Это может быть полезно при анализе предпочтений покупателей или оценке различных характеристик домов в зависимости от количества этажей.

### Диаграмма рассеяния (scatter plot): Позволяет исследовать связь между двумя переменными.

```python
import matplotlib.pyplot as plt

plt.scatter(data['sqft_living'], data['price'])
plt.xlabel('Жилая площадь (кв. футы)')
plt.ylabel('Стоимость недвижимости ($)')
plt.title('Зависимость стоимости от жилой площади')
plt.show()
```

Вывод: По графику scatter plot можно заметить, что существует положительная связь между площадью жилой площади и стоимостью недвижимости. Точки на графике показывают, что при увеличении площади жилой площади стоимость недвижимости также увеличивается. Однако стоит отметить, что присутствуют некоторые выбросы, где некоторые недвижимости имеют высокую стоимость при небольшой площади жилой площади. Эти выбросы могут быть обусловлены другими факторами, такими как расположение недвижимости или наличие особых особенностей. В целом, график подтверждает гипотезу о том, что площадь жилой площади является важным фактором, влияющим на стоимость недвижимости.







### Диаграмма ящика с усами (box plot): Позволяет визуализировать распределение переменной и выявить выбросы. 

```python
import matplotlib.pyplot as plt

# Группировка данных по количеству спален и получение стоимости недвижимости
grouped_data = data.groupby('bedrooms')['price'].apply(list)

# Построение box plot
plt.boxplot(grouped_data)
plt.xlabel('Количество спален')
plt.ylabel('Стоимость недвижимости ($)')
plt.title('Распределение стоимости недвижимости по количеству спален')
plt.xticks(range(1, len(grouped_data) + 1), grouped_data.index)
plt.show()

```

Вывод: График box plot позволяет сравнить распределение стоимости недвижимости для разных количеств спален. Мы видим, что с увеличением количества спален обычно наблюдается увеличение медианной стоимости недвижимости. Однако стоит отметить, что есть некоторые выбросы, особенно для недвижимости с большим количеством спален. Например, некоторые дома с 7 или более спальнями имеют очень высокую стоимость. В целом, график подтверждает, что количество спален может влиять на стоимость недвижимости, но также есть и другие факторы, которые могут вносить значительное влияние на цену.

### Гистограмма (histogram): Позволяет оценить распределение переменной.

```python
import matplotlib.pyplot as plt

# Построение гистограммы
plt.hist(data['price'], bins=20)
plt.xlabel('Стоимость недвижимости ($)')
plt.ylabel('Частота')
plt.title('Распределение стоимости недвижимости')
plt.show()
```

Вывод: Гистограмма позволяет нам получить представление о распределении стоимости недвижимости. Мы видим, что большинство наблюдений сосредоточено в нижних диапазонах стоимости, а с увеличением стоимости количество наблюдений уменьшается. Это говорит о том, что большинство недвижимости имеет относительно низкую цену, а недвижимость с более высокой стоимостью является более редким явлением. Эта информация может быть полезной при анализе и принятии решений о стоимости недвижимости.

### Круговая диаграмма (pie chart): Позволяет визуализировать доли категориальных переменных.

```python
import matplotlib.pyplot as plt

# Получение данных
room_counts = data['bedrooms'].value_counts()

# Построение круговой диаграммы
plt.pie(room_counts, labels=room_counts.index, autopct='%1.1f%%')
plt.title('Распределение количества комнат')
plt.axis('equal')
plt.show()
```

Вывод: Из круговой диаграммы видно, что большинство наблюдений (более 60%) имеют 3 или 4 комнаты. Количество недвижимости с другими количествами комнат значительно меньше. Эта информация может быть полезна при анализе спроса на недвижимость с разным количеством комнат и принятии решений о разработке и маркетинге недвижимости.

### Линейная диаграмма (line plot): Позволяет отслеживать изменение переменной во времени.

```python
import matplotlib.pyplot as plt

# Группировка данных по году постройки и вычисление средней цены
mean_price_by_year = data.groupby('yr_built')['price'].mean()

# Построение графика линии
plt.plot(mean_price_by_year.index, mean_price_by_year.values)
plt.title('Средняя цена недвижимости в зависимости от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Средняя цена')
plt.show()
```

Вывод: Из графика видно, что средняя цена недвижимости имеет различную динамику в зависимости от года постройки. Наблюдаются колебания цен в разные годы, возможно, связанные с факторами, влияющими на спрос и предложение на рынке недвижимости. Анализ таких изменений может помочь в понимании цикличности рынка и принятии решений о времени покупки или продажи недвижимости.
